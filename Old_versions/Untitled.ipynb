```python
nt = 15000 #15000
r_bar = 100000
k = 0.05
s2pv = 100000000
s2s = 150000000
R = 2000
rho = 0.0006
l = 0.075 # an ZI agent arrives per 13 to 14 time steps
```

```python
spv = 10000
ss = s2s**0.5 #12247.4
```

```python
np.random.seed(12345)
ZI_agent_arrival = np.random.exponential(1/l,nt)
u = np.random.normal(0,ss,nt)
```

```python
r = [r_bar]
for t in range(1,nt):
    # update price
    r.append(max(0,k*r_bar+(1-k)*r[t-1]+u[t]))
```

```python
plt.figure(figsize=(10,6))
plt.plot(r)
plt.title("Price Process", fontsize=18)
plt.xlabel("time(milliseconds)", fontsize=15)
plt.ylabel("r", fontsize=15)
plt.show()
```
